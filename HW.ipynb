# Tesla Stock Analysis
Analysis of 10 years of Tesla stock data with volatility and returns analysis

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Tesla ticker
ticker = "TSLA"

# Download 10 years of daily data
df = yf.download(ticker, period="10y", interval="1d", auto_adjust=True)

# Clean data
df = df.dropna()

# Calculate daily returns
df["Returns"] = df["Close"].pct_change()

# Calculate annualized volatility (252 trading days)
df["Annualized Volatility"] = df["Returns"].rolling(window=252).std() * np.sqrt(252)
```

```python
# 1. Plot stock prices
plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Close"])
plt.title("Tesla (TSLA) Stock Prices - Last 10 Years")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
# 2. Plot stock returns
plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Returns"])
plt.title("Tesla (TSLA) Daily Stock Returns")
plt.xlabel("Date")
plt.ylabel("Return")
plt.grid(True)
plt.show()
```

```python
# 3. Plot annualized volatility
plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Annualized Volatility"])
plt.title("Tesla (TSLA) Annualized Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.grid(True)
plt.show()
```
